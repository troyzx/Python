```python
class 大象():
    def 打招呼():
        return "I'm an elephant"
    def 吃():
        return "yammy"
    def 睡():
        return "zzzzz"

class 冰箱():
    def __init__(self):
        self.door=False
        self.items=[]
    def 开门(self):
        self.door=True
        print('门开了')
    def 关门(self):
        self.door=False
        print('门关了')
    def 放(self, 物品):
        self.items.append(物品)
```
