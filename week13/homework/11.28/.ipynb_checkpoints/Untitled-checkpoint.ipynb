```python
import numpy as np
from matplotlib import pyplot
```

```python
blackhole = [0,0]

def F(pos):
    x = pos[0]
    y = pos[1]
    r2 = x**2+y**2
    if r2==0:
        return 0,0
    theta = np.arccos(x/r2**0.5)
    F = 10*r2**0.5
    fx = abs(F*np.cos(theta))*np.sign(x)
    fy = abs(F*np.sin(theta))*np.sign(y)
    return fx, fy
    
class dot():
    
    def __init__(self, pos_0 = [0,0]):
        self.pos = pos_0
        
    def walk(self, n=1):
        step = np.random.random(n)
        theta = np.random.random(n)*2*np.pi
        dx = np.dot(np.cos(theta), step)
        dy = np.dot(np.sin(theta), step)
        self.pos += np.array([dx,dy])
```

```python
plt.figure(figsize = (30,30))
pos = []
for n in range(10000):
    i = np.random.randint(-50,50)
    j = np.random.randint(-50,50)
    dot1 = dot([i ,j])
    dot1.walk(10000)
    plt.scatter(dot1.pos[0], dot1.pos[1])
    
plt.show()
```
