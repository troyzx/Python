```python
from tqdm import tqdm
import numpy as np
from matplotlib import pyplot as plt
 

def F(pos):
    x,y = pos
    r = (x**2+y**2)**0.5
    if r==0:
        return (0,0)
    theta = np.arccos(pos[0] / r)
    dx = abs(5 * np.cos(theta))/r**2 * np.sign(x)
    dy = abs(5 * np.sin(theta))/r**2 * np.sign(y)
    return (dx,dy)


class Singledot():

    def __init__(self, pos0=[0,0]):
        self.pos = pos0
        
    def walk(self, n=1):
        x = self.pos[0]
        y = self.pos[1]
        theta = np.array([np.random.random(n)*2*np.pi])
        step = np.array([np.random.random(n)])
        steptheta = np.concatenate((theta.T, step.T), axis = 1)
        for step, theta in steptheta:
            dx =  F([x,y])[0] + np.cos(theta)*step
            dy =  F([x,y])[1] + np.sin(theta)*step
            x += dx
            y += dy
        self.pos = np.array([x,y])
```

```python
plt.figure(figsize=(8,8))
for i in tqdm(range(100)):
    point = Singledot([np.random.randint(-5,5), np.random.randint(-5,5)])
    point.walk(100)
    plt.scatter(point.pos[0], point.pos[1])
plt.show()
```
